```python
import pandas as pd
data=pd.read_csv('fruit_data_with_colours.csv')
data.head()
```

```python
dff=data.drop('fruit_label',axis=1)
df=dff.drop('fruit_subtype',axis=1)
X=df.drop('fruit_name',axis=1)
y=data['fruit_label']
```

```python
y
```

```python
X
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
#feature selection using chi-square test
k_selected_features=4
chi2_selector=SelectKBest(chi2,k=k_selected_features)
X_selected=chi2_selector.fit_transform(X,y)
```

```python
#split to training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected,y,test_size=0.2,random_state=42)
```

```python
#train the SVM classifier
from sklearn.svm import SVC
svm_classifier=SVC(kernel='linear')
svm_classifier.fit(X_train,y_train)
```

```python
#evaluate svm classifier
from sklearn.metrics import accuracy_score,classification_report
y_pred=svm_classifier.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
```

```python
#print the results
print("Accuracy::",accuracy)
print("Classification Report:: ")
print(report)
```

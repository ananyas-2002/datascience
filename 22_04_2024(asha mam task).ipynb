```python
def remove_urls(text):
    return re.sub(r'https?://\S+|www\.S+', '', text)

# Apply the remove_urls function to the 'clean_text' column
df['clean_url_text'] = df['text'].apply(remove_urls)

# Display the DataFrame with cleaned text
```

```python
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords data
nltk.download('stopwords')

nltk.download('punkt')
# Set of English stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from a single text
def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

# Apply the remove_stopwords function to a DataFrame column containing text
df['clean_stop_text'] = df['clean_url_text'].apply(remove_stopwords)
```

```python
df
```

```python
import re

# Function to remove digits from a single text
def remove_digits(text):
    return re.sub(r'\d+', '', str(text))
# Apply the remove_digits function to a DataFrame column containing text
df['clean_digit_text'] = df['clean_stop_text'].apply(remove_digits)
```

```python
df
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('punkt')

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Function to perform stemming on a single text
def perform_stemming(text):
    # Tokenize the text
    words = word_tokenize(text)
    
    # Apply stemming to each word
    stemmed_words = [stemmer.stem(word) for word in words]
    
    # Join the stemmed words back into a single string
    stemmed_text = ' '.join(stemmed_words)
    
    return stemmed_text

# Apply the perform_stemming function to the 'clean_url_text' column
df['clean_stemmed_text'] = df['clean_digit_text'].apply(perform_stemming)
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# Function to perform lemmatization on a single text
def perform_lemmatization(text):
    # Tokenize the text
    words = word_tokenize(text)
    
    # Apply lemmatization to each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    
    # Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_words)
    
    return lemmatized_text

# Apply the perform_lemmatization function to the 'clean_digit_text' column
df['clean_lemmatized_text'] = df['clean_digit_text'].apply(perform_lemmatization)
```

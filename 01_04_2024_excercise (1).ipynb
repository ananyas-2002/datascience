```python
import librosa
import IPython.display as ipd

# Load an audio file
file_path = "C:/Users/admin/DataScience Lab/sample.wav"
audio_data, sample_rate = librosa.load(file_path, sr=None)

# Create an audio widget to play the audio data
audio_widget = ipd.Audio(data=audio_data, rate=sample_rate)

# Display the audio widget
display(audio_widget)
```

```python
# read video file
import cv2 
import numpy as np 
cap = cv2.VideoCapture("C:/Users/admin/DataScience Lab/video.mp4") 
if (cap.isOpened()== False): 
    print("Error opening video file") 
while(cap.isOpened()): 
    ret, frame = cap.read() 
    if ret == True:
        cv2.imshow('Frame', frame) 
        if cv2.waitKey(25)& 0xFF == ord('q'): 
            break
    else: 
        break
cap.release() 
cv2.destroyAllWindows() 
```

```python
import cv2
video_path ="C:/Users/admin/DataScience Lab/video1.mp4"
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()
while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('Video', frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

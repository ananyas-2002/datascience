```python
#range without built-in function
lst=[]
n=int(input("Enter number of elements::"))
for i in range(0,n):
    ele=int(input("Enter elements::"))
    lst.append(ele)
print("List::",lst)
mindata=maxdata=lst[0]
for ele in lst:
    if ele<=mindata:
        mindata=ele
    if ele>=maxdata:
        maxdata=ele
range_value=maxdata-mindata
print("Range(without function)::",range_value)

#range using built-in function
import numpy as np
max_data=np.max(lst)
min_data=np.min(lst)
range_value=max_data-min_data
print("Range(using built-in function)::",range_value)
```

```python
#variance without built-in function
lst=[]
n=int(input("Enter number of elements::"))
for i in range(0,n):
    ele=int(input("Enter elements::"))
    lst.append(ele)
print("List::",lst)
total=0
for i in lst:
    total+=i
mean=total/len(lst)
sum_value=0
for item in lst:
    sum_value+=(item-mean)**2
variance=sum_value/len(lst)
print("Variance::",variance)

#variance using built-in function
import numpy as np
print("Variance(using function)::",np.var(lst))
```

```python
#standard deviation without built-in function
import math
lst=[]
n=int(input("Enter number of elements::"))
for i in range(0,n):
    ele=int(input("Enter elements::"))
    lst.append(ele)
print("List::",lst)
total=0
for i in lst:
    total+=i
mean=total/len(lst)
sum_value=0
for item in lst:
    sum_value+=(item-mean)**2
variance=sum_value/len(lst)
print("Variance::",variance)
standard_deviation=math.sqrt(variance)
print("Standard deviation(without using function)::",standard_deviation)

#standard deviation using built-in function
import numpy as np
print("Standard deviation::",np.std(lst))
```

```python
#IQR without using built-function
from scipy import stats
def median(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    if n % 2 == 0:
        return (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2
    else:
        return sorted_arr[n // 2]

def quartiles(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    q2 = median(sorted_arr)
   
    if n % 2 == 0:
        lower_half = sorted_arr[:n // 2]
        upper_half = sorted_arr[n // 2:]
    else:
        lower_half = sorted_arr[:n // 2+1]
        upper_half = sorted_arr[n // 2 + 1:]

    q1 = median(lower_half)
    q3 = median(upper_half)

    return q1, q2, q3

def iqr(arr):
    q1, q2, q3 = quartiles(arr)
    return q3 - q1
lst=[]
n=int(input("Enter number of elements::"))
for i in range(0,n):
    ele=int(input("Enter elements::"))
    lst.append(ele)
print("List::",lst)
print("IQR(without built-in):", iqr(lst))
#IQR using built-in function
IQR=stats.iqr(lst,interpolation='midpoint')
print("IQR::",IQR)
```
